```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import ast
```

```python
df = pd.read_csv("Music_all.csv")
```

# Creating predictable dataset

```python
data_dum = pd.get_dummies(df, columns = ["Chords"])
```

```python
data_dum
```

```python
df
```

```python
pred_df = pd.DataFrame(columns = ["Genre", "Chord1", "Chord2", "Chord3", "Chord4", "Chord5"] )
for i in range (df.shape[0]):
    name = df.Genre.loc[i]
    a = df.Chords.loc[i]
    a = ast.literal_eval(a)
    a = [n.strip() for n in a]
    if len(a) >= 5:
        pred_df = pred_df.append({"Genre" : name, "Chord1" : a[0], "Chord2": a[1], "Chord3" : a[2], "Chord4" : a[3], "Chord5" : a[4]}, ignore_index = True)
    if len(a) == 4:
        pred_df = pred_df.append({"Genre" : name, "Chord1" : a[0], "Chord2": a[1], "Chord3" : a[2], "Chord4" : a[3], "Chord5" : a[0]}, ignore_index = True)
    if len(a) == 3:
        pred_df = pred_df.append({"Genre" : name, "Chord1" : a[0], "Chord2": a[1], "Chord3" : a[2], "Chord4" : a[0], "Chord5" : a[0]}, ignore_index = True)
    if len(a) == 2:
        pred_df = pred_df.append({"Genre" : name, "Chord1" : a[0], "Chord2": a[1], "Chord3" : a[0], "Chord4" : a[0], "Chord5" : a[0]}, ignore_index = True)
    if len(a) == 1:
        pred_df = pred_df.append({"Genre" : name, "Chord1" : a[0], "Chord2": a[0], "Chord3" : a[0], "Chord4" : a[0], "Chord5" : a[0]}, ignore_index = True)
    
```

```python
data_dum = pd.get_dummies(pred_df, columns = ["Chord1", "Chord2", "Chord3", "Chord4", "Chord5"])
```

```python
data_dum
```

# Predicting model (model and shuffling)

```python
pred_data = data_dum.copy()
df_shuf = pred_data.sample(frac=1).reset_index(drop=True)
```

```python
train, test = train_test_split(df_shuf, test_size = 0.2)
test1 = test.drop(columns = "Genre")
test.reset_index(inplace = True)
```

```python
X_train = train.drop(columns = "Genre")
Y_train = train.Genre
```

## KNN

```python
model = KNeighborsClassifier(n_neighbors = 1).fit(X_train, Y_train)
```

```python
KNN_pred = model.predict(test1)
```

```python
def score(arr, data):
    blues = 0;
    classical = 0;
    country = 0;
    disco = 0;
    hiphop = 0;
    jazz = 0;
    metal = 0;
    pop = 0;
    reggae = 0;
    rock = 0;
    correct = 0
    for i in range(len(arr)):
            if (arr[i] == data.Genre.loc[i]):
                correct +=1
                if(arr[i] == "Blues"):
                    blues += 1
                if(arr[i] == "Classical"):
                    classical += 1
                if(arr[i] == "Country"):
                    country += 1
                if(arr[i] == "Disco"):
                    disco += 1
                if(arr[i] == "Hiphop"):
                    hiphop += 1
                if(arr[i] == "Jazz"):
                    jazz += 1
                if(arr[i] == "Metal"):
                    metal += 1
                if(arr[i] == "Pop"):
                    pop += 1
                if(arr[i] == "Reggae"):
                    reggae += 1
                if(arr[i] == "Rock"):
                    rock += 1 
    print("Blues: " + str(blues))
    print("Classical: " + str(classical))
    print("Country: " + str(country))
    print("Disco: " + str(disco))
    print("Hiphop: " + str(hiphop))
    print("Jazz: " + str(jazz))
    print("Metal: " + str(metal))
    print("Pop: " + str(pop))
    print("Reggae: " + str(reggae))
    print("Rock: " + str(rock))
    return correct/len(arr)*100
```

```python
score(KNN_pred, test)
```

## Random Forest Classifier

```python
model_forest = RandomForestClassifier(n_estimators = 100, random_state = 0).fit(X_train, Y_train)
```

```python
Forest_pred = model_forest.predict(test1)
```

```python
score(Forest_pred, test)
```
